```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from pandas.io.parsers import read_csv
import warnings
warnings.filterwarnings("ignore")
from matplotlib.pyplot import figure
```

```python
PATH = "data/Happiness/"
```

```python
df2015 = pd.read_csv(f'{PATH}2015.csv', low_memory=False)
df2016 = pd.read_csv(f'{PATH}2016.csv', low_memory=False)
df2017 = pd.read_csv(f'{PATH}2017.csv', low_memory=False)
df2018 = pd.read_csv(f'{PATH}2018.csv', low_memory=False)
df2019 = pd.read_csv(f'{PATH}2019.csv', low_memory=False)
```
